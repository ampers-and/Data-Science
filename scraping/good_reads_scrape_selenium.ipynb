```python
driver = webdriver.Chrome('C:/Users/Cactuar/Downloads/chromedriver_win32/chromedriver.exe', options=options)
#driver.set_window_position(-2000,0)
first_url = 1
last_url = 140    # Last book is 8,630,000
#count = 0
#reqs = ["https://stackoverflow.com/questions/46"]

slp = time.sleep(1)
timeout = 5

for book_reference_number in range(first_url, last_url):
    #try_page = lambda x: return driver.get("https://www.goodreads.com/book/show/"+str(book_reference_number))
    
    #count += 1
    #if count%60 == 0:
    #    driver.quit()
    #    time.sleep(20)
    #    driver = webdriver.Chrome('C:/Users/Cactuar/Downloads/chromedriver_win32/chromedriver.exe', options=options)
    
    #book_reference_number = np.random.randint(1,8500000)
    
    #driver.get(choice(reqs)+str(np.random.randint(1,10000)))   
    
    
    
    driver.get("https://www.goodreads.com/book/show/"+str(book_reference_number))
    
    try:
        element_present = EC.presence_of_element_located((By.ID, 'element_id'))
        WebDriverWait(driver, timeout).until(element_present)
        time.sleep(30)
        driver.navigate().refresh()
    except TimeoutException as err:
        print(err)
    except NoSuchElementException as ex:
        print('ex')
        print('caught')
        time.sleep(7)
        driver.navigate().refresh()
        
    
    
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    
    
    try:
        book_title = soup.select('.gr-h1.gr-h1--serif')[0].text.strip()
    except NoSuchElementException as err:
        print(err)
        print('caught1')
    except:
        book_title = 'not found'
    
    
    
    try:
        author_name = soup.select('.authorName')[0].text.strip()
    except:
        author_name = 'not found'
        
    
    
    try:
        description = soup.select('.readable.stacked')[0].text.strip()
    except:
        description = 'not found'

       
    
    try:
        rating = soup.find("span", itemprop="ratingValue").text.strip()
    except:
        rating = 'not found'
    
    
    
    try:
        total_ratings = soup.find(itemprop="ratingCount").get("content")
    except:
        total_ratings = 'not found'
       
    
    
    try:
        total_reviews = soup.find(itemprop="reviewCount").get("content")
    except:
        total_reviews= 'not found'
    
    
    
    try:
        num_pages = soup.find("span", itemprop="numberOfPages").text.strip()[:-5]
    except:
        num_pages = 'not found'
    
    
    
    try:
        published = soup.find_all("div", {"class":"row"})[1].text.strip()
    except:
        published = 'not found'
    
    
    
    try:
        ISBN = soup.find_all("div", {"class":"infoBoxRowItem"})[1].text.strip()
    except:
        ISBN ='not found'
    
    try:
        genre_list = soup.find_all('div', class_='elementList')
        genres = []
        for genre in genre_list:
            genre.text.split()
            genres.append(genre.text.split()[0])
    except:
        genre_list = 'not found'
    
    
    
    
    
    good_reads_df = good_reads_df.append(
        {
            'title': book_title,
            'author': author_name,
            'description': description,
            'aggregate_rating': rating,
            'total_ratings': total_ratings,
            'total_reviews': total_reviews,
            'num_pages': num_pages,
            'publish_date': published,
            'ISBN': ISBN,
            'genres' : genres
        },
        ignore_index=True
    
    )
```
