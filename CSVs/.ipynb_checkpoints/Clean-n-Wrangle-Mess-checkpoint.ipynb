```python
df = pd.read_csv('bb_details.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['descrip'], inplace=True)
```

```python
for n in df.n:
    df.replace(n, n.split('-')[0], inplace=True)
```

```python
df.head()
```

```python
def strip_extra(df, col):
    for val in df[col]:
        try:
            new_val = val.strip()
            df.replace(val, new_val, inplace=True)
        except Exception:
            continue
    return df

def clean_line_breaks(df, col):
    for val in df[col]:
        try:
            val_l = val.split('\n  ')
            new_val = ' '.join(val_l)
            df.replace(val, new_val, inplace=True)
        except Exception:
            continue
    return df

def remove_paren(df, col):
    for val in df[col]:
        try:
            new_val = val.strip(')')
            df.replace(val, new_val, inplace=True)
        except Exception:
            continue
```

```python
cols = ['title', 'rating', 'num_ratings', 'num_reviews', 'published_on']

for col in cols:
    strip_extra(df, col)
```

```python
df.head()
```

```python
cols = ['num_ratings', 'num_reviews', 'published_on']

for col in cols:
    clean_line_breaks(df, col)
```

```python
remove_paren(df, 'isbn13')
```

```python
df.head()
```
