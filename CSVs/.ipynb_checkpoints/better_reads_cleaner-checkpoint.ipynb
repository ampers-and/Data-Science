```python
def strip_extra(df, col):
    '''Cleans extra whitespace from end of strings'''
    for val in df[col]:
        try:
            new_val = val.strip()
            df.replace(val, new_val, inplace=True)
        except Exception:
            continue
    return df

def clean_line_breaks(df, col, break_on='\n  '):
    '''After stripping white space, cleans line breaks that still remain'''
    for val in df[col]:
        try:
            val_l = val.split(break_on)
            new_val = ' '.join(val_l)
            df.replace(val, new_val, inplace=True)
        except Exception:
            continue
    return df

def remove_paren(df, col):
    '''Get rid of trailing paren in ISBN-13 col'''
    for val in df[col]:
        try:
            new_val = val.strip(')')
            df.replace(val, new_val, inplace=True)
        except Exception:
            continue
    return df
            
def find_bad_rows(df, col='n'):
    '''Finds values that are not numbers in webpage column'''
    df_X = df.copy()
    string_list = ['n', 'reviews', 'review', 'ratings', 'rating']
    for string in string_list:
        mask = df_X[col].str.contains(string)
        df_X =  df_X[~mask]
    
    return df_X

def impute_isbn(df, isbn_df):
    '''Impute the missing ISBN from dataframe of the ISBNs using webpage id'''
    for idx, row in df.iterrows():
        if int(row['n']) in missing_isbn['n'].values:
            missing_idx = missing_isbn[missing_isbn['n'] == int(row['n'])].index.values[0]
            row['isbn'] = missing_isbn.iloc[missing_idx]['isbn']
            row['isbn13'] = missing_isbn.iloc[missing_idx]['isbn13']
    return df
```
